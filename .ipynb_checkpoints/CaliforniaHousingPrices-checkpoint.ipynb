# Create Train & Test Set

```python
# Define Function to split train and test
def split_train_test(data, test_ratio):
    np.random.seed(40)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[0 : test_set_size]
    train_indices = shuffled_indices[test_set_size : ]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(df, 0.2)
```

```python
print(len(train_set))
print(len(test_set))
```

```python
# Create a hash of each instance's identifier and put that instance in the test set if the hash is lower or equal to 20%
# of the maximum hash value. This ensures that the test set will remain consistent across multiple runs, even if you
# refresh the dataset.

# But in order to implement the above concept we need to have a identity column OR we can also create an identity column from existing columns.
```

```python
# Example
from zlib import crc32
n = 3
print(crc32(np.int64(n)) & 0xffffffff)
print((crc32(np.int64(n)) & 0xffffffff) < (0.2 * 2**32))
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins = [0, 1.5, 3, 4.5, 6, np.inf], labels = [1, 2, 3, 4, 5])
df['income_cat'].hist()
```

```python
# Stratified Shuffle
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index] 
```

```python
# Drop the attribute
```

```python
strat_train_set.drop('income_cat', axis = 1, inplace = True)
strat_test_set.drop('income_cat', axis = 1, inplace = True)
```
